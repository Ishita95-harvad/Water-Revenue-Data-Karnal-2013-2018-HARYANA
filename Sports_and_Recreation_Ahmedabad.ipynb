## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and potential outliers.


**Reasoning**:
Set 'Sr. No.' as index, clean 'Phone No. / Timings' and 'Address', and standardize text columns.



```python
# Set 'Sr. No.' as index
df = df.set_index('Sr. No.')

# Clean 'Phone No. / Timings'
import re

def clean_phone_number(text):
    if pd.isna(text):
        return None
    phone_match = re.search(r'\d{2,3}-\d{7,8}', text)
    if phone_match:
        return phone_match.group(0)
    return None

def clean_timing(text):
    if pd.isna(text):
        return None
    timing_match = re.search(r'[a-zA-Z0-9\s.:-]+(?:A\.M\.|P\.M\.)', text)
    if timing_match:
        return timing_match.group(0)
    return None

df['Phone No.'] = df['Phone No. / Timings'].apply(clean_phone_number)
df['Timings'] = df['Phone No. / Timings'].apply(clean_timing)
df = df.drop('Phone No. / Timings', axis=1)

# Clean 'Address'
df['Address'] = df['Address'].str.strip()
df['Address'] = df['Address'].str.title()

# Standardize text columns
text_cols = ['Name of the Sports Complex', 'Area', 'Address', 'Category']
for col in text_cols:
    df[col] = df[col].str.strip()
    df[col] = df[col].str.title()
display(df)
```

## Feature engineering

### Subtask:
Engineer new features from the cleaned dataset to potentially improve model performance.


**Reasoning**:
Apply one-hot encoding to categorical features, and then scale numerical features using MinMaxScaler.



**Reasoning**:
The error `InvalidIndexError: Reindexing only valid with uniquely valued Index objects` suggests that the index of the dataframe `df` is not unique.  Reset the index before concatenation.



```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import pandas as pd
import numpy as np

# Reset the index of df before concatenation
df = df.reset_index()

# One-hot encode 'Area' and 'Category'
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_features = ohe.fit_transform(df[['Area', 'Category']])
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names_out(['Area', 'Category']))
df = pd.concat([df, encoded_df], axis=1)

# Drop original columns and 'Sr. No.' after one-hot encoding
df = df.drop(['Area', 'Category', 'Sr. No.'], axis=1)

# Identify numerical features for scaling
numerical_cols = ['Phone No.', 'Timings']
for col in numerical_cols:
  if df[col].dtype == 'object':
    df[col] = df[col].astype(str).replace('None', np.nan)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].mean())
    # Scale only numerical columns after converting them to numeric and filling NaNs
    scaler = MinMaxScaler()
    df[col] = scaler.fit_transform(df[[col]])

display(df)
```

**Reasoning**:
The warning "All-NaN slice encountered" indicates that the 'Phone No.' and 'Timings' columns contain only NaN values after cleaning.  Since these columns are now useless for numerical feature scaling, they should be dropped instead of attempting to scale them.



```python
# Drop 'Phone No.' and 'Timings' columns
df = df.drop(['Phone No.', 'Timings'], axis=1)
display(df)
```

## Model optimization

### Subtask:
Optimize the hyperparameters of the Logistic Regression model to improve its performance.


**Reasoning**:
I need to optimize the hyperparameters of the Logistic Regression model using GridSearchCV.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the parameter grid
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'],  # Solvers that support both l1 and l2 penalties
    'max_iter': [100, 500, 1000]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=2, scoring='accuracy') # Using 2 folds due to small dataset

# Fit GridSearchCV to the training data
grid_search.fit(X_train_numeric, y_train)

# Get the best hyperparameters and best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print(f"Best Hyperparameters: {best_params}")

# Evaluate the best model on the validation set
y_pred_best = best_model.predict(X_val_numeric)
best_accuracy = accuracy_score(y_val, y_pred_best)
print(f"Validation Accuracy with Best Model: {best_accuracy}")
```
