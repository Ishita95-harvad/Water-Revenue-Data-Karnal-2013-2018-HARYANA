## Data visualization

### Subtask:
Visualize the key findings from the data analysis, acknowledging the limitations of the single data point.


**Reasoning**:
Create visualizations to show the limitations of the single data point, including a bar chart of missing values and bar charts for categorical columns. Add a summary section to describe the visualizations and insights (or lack thereof).



```python
import matplotlib.pyplot as plt

# Missing Value Visualization
plt.figure(figsize=(12, 6))
missing_values = df.isnull().sum()
plt.bar(missing_values.index, missing_values.values, color='skyblue')
plt.xlabel("Columns", fontsize=12)
plt.ylabel("Number of Missing Values", fontsize=12)
plt.title("Missing Values per Column", fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Categorical data visualization
categorical_columns = ['City Name', 'Zone Name', 'Ward Name']
plt.figure(figsize=(15, 5))
for i, col in enumerate(categorical_columns):
    plt.subplot(1, len(categorical_columns), i + 1)
    if df[col].notnull().any():  # Check if the column is not entirely NaN
      df[col].value_counts().plot(kind='bar', color=['lightcoral', 'lightgreen', 'lightsteelblue'])
      plt.xlabel(col, fontsize=10)
      plt.ylabel("Count", fontsize=10)
      plt.title(f"Unique Values in '{col}'", fontsize=12)
      plt.xticks(rotation=45, ha='right')
      plt.tight_layout()
    else:
      plt.text(0.5, 0.5, f"All values in '{col}' are missing", ha='center', va='center')
      plt.xticks([])
      plt.yticks([])
      plt.title(f"Unique Values in '{col}'", fontsize=12)


plt.show()


print("Visualization Summary:")
print("Due to the single data point, the visualization does not represent a true distribution. Standard methods like histograms and boxplots are not meaningful for one datapoint.")
print("The bar chart showing missing values gives a clear picture of the missing values per column, but is not a statistical distribution in this context.")
print("Bar plots created for categorical columns show the unique value and its counts.  Again, these cannot represent distributions.")
print("More data is needed to perform meaningful analysis.")
```
